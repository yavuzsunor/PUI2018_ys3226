### Fit the data with a line in log space:

```python
OLS1 = smf.ols(formula='logNrgPerBuild  ~ logUnit', data=bblnrgdataGPD).fit()
OLS1.summary()
```

```python
OLS2 = smf.ols(formula='logUnit ~ logNrgPerBuild', data=bblnrgdataGPD).fit()
OLS2.summary()
```

```python
fig = plt.figure(figsize = (15,7))

ax = fig.add_subplot(121)

ax.scatter(bblnrgdataGPD['UnitsTotal'], bblnrgdataGPD['NrgperBuild'])
ax.plot(bblnrgdataGPD['UnitsTotal'], e**OLS1.predict(bblnrgdataGPD), color='black')
ax.set_yscale('log')
ax.set_xscale('log')
ylim(1e2,1e11)
xlim(1,1e4)
plt.title('a) Energy cons.(depend) vs. Number of units(independ) in log-scale with a linear model', 
          fontsize=10)
plt.xlabel('Number of Units in Building', fontsize=10)
plt.ylabel('Energy consumption per building', fontsize=10)

ax = fig.add_subplot(122)

ax.scatter(bblnrgdataGPD['NrgperBuild'], bblnrgdataGPD['UnitsTotal'])
ax.plot(bblnrgdataGPD['NrgperBuild'], e**OLS2.predict(bblnrgdataGPD),  color='black')
ax.set_yscale('log')
ax.set_xscale('log')
xlim(1e2,1e11)
ylim(1,1e4)
plt.title('b) Number of units(depend) vs. Energy cons.(independ) in log-scale with a linear model', 
          fontsize=10)
plt.xlabel('Energy consumption per building', fontsize=10)
plt.ylabel('Number of Units in Building', fontsize=10)

pl.show()
```

Figure 6: Number of units vs Energy consumption as a scatterplot with log-scale and the linear fit using:

a) number of units as independent and energy as dependent variable.

b) number of units as dependent and energy as independent variable.

### Second degree polynomial fit

```python
# create a new feature for logUnit^2
bblnrgdataGPD['logUnit2'] = bblnrgdataGPD['logUnit']**2
```

```python
OLS3 = smf.ols(formula='logNrgPerBuild  ~ logUnit2 + logUnit', data=bblnrgdataGPD).fit()
OLS3.summary()
```

```python
fig = plt.figure(figsize = (20,7))

ax = fig.add_subplot(121)

ax.scatter(bblnrgdataGPD['UnitsTotal'], bblnrgdataGPD['NrgperBuild'])
ax.scatter(bblnrgdataGPD['UnitsTotal'], e**OLS3.predict(bblnrgdataGPD), color='black')
ax.set_yscale('log')
ax.set_xscale('log')
ylim(1e2,1e11)
xlim(1,1e4)
plt.title('Energy cons.(depend) vs. Number of units(independ) in log-scale with a quadratic model', 
          fontsize=10)
plt.xlabel('Number of Units in Building', fontsize=10)
plt.ylabel('Energy consumption per building', fontsize=10)

plt.show()
```

### Define a chisquare function

```python
## create a predict column to be used in chi2 function
bblnrgdataGPD['PredNrg'] = OLS3.predict()
```

```python
## define chi2 function to calculate chi2
def chi2(data, model, errors = None):
    '''Calculates the chi sq given data, model and errors
    Arguments:
    data: series of datapoints (endogenous variable)
    model: series of predicted values corresponding to the observed data
    errors: serie of errors (optional). 
    If errors are not passes all errors are set to 1
    '''
    
    np.random.randn(123)
    chi2 = 0 
    if errors is None:
        for index in bblnrgdataGPD.index:
            chi2 += (model[index] - data[index])**2 / 1
    else:
        for index in bblnrgdataGPD.index:
            chi2 += (model[index] - data[index])**2 / errors
    
    return chi2
```
