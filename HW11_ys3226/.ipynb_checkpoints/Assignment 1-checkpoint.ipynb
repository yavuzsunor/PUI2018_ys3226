## 2: finding trends

```python
mta_trend = np.nansum(mta, axis=0)
```

```python
mta_trend.shape[0]
```

```python
fig = plt.figure(figsize = (15,7))

for i in range(mta_trend.shape[0]):
    plt.plot(np.log10(mta_trend[i,:]))

plt.xlabel('weeks since 05/21/2010', fontsize=20)
plt.ylabel('swipes (log10)', fontsize=20)
plt.show()    
```

            Figure 4: Checking with eye for the steepest increase and decrease of each ridership types.

```python
def changeFinder(data):

    n = len(data)
    datamean = data.mean()
    datasqmean = (data**2).mean()
    fac = datasqmean - datamean**2
    
    datacsum = data.cumsum()
    datasum = datacsum[-1]

    ppoc = np.zeros(n) #container for point of change relative prob

    #online (iterative) search for point of change
    for m in range(n-1):
        pos = m + 1
        relativePosition = (pos) * (n - pos)
 
        Q = datacsum[m] - (datasum - datacsum[m]) #cumsum up to m - cumsum after
        
        U = -(datamean * (n - 2 * pos) + Q)**2 / (4.0 * relativePosition) + fac
       
        ppoc[m+1] = (-(n * 0.5 - 1) * np.log(n * U * 0.5) - 
                 0.5 * np.log(relativePosition))

    ppoc[0] = min(ppoc[1:])
    changePoint = np.argmax(ppoc)
    
    return {'pChange': ppoc, 
            'pointOfChange': changePoint + 1,
            'meanBefore': (data[:changePoint+1]).mean(), 
            'meanAfter': (data[(changePoint+1):]).mean()}
```

```python
for i in range(mta_trend.shape[0]):
    pochange = changeFinder(np.log10(mta_trend[i,:]))
    print("PoC for ridership", i+1,  pochange['pointOfChange'], "from", 
          np.log10(mta_trend[i,pochange['pointOfChange']-2]), "to", 
          np.log10(mta_trend[i,pochange['pointOfChange']-1]), "with a change of",
          np.log10(mta_trend[i,pochange['pointOfChange']-1]) / np.log10(mta_trend[i,pochange['pointOfChange']-2]) -1 
         )
```

## Comment: 

### The table above can guide us to say steepest increase(it's obviously 23) and decrease, however when we look at the plot by eye, we actually see that the steepest increase and decrease occured in the same ridership. That's why we should check the time-series one more time after partioning it.

### However, the changeFinder function didnot work when I partioned the timeseries. It didnot capture the big(and the steepest decrease) at the Ridership number 23.

### So, I basically created my own function as a double for loop and put the each increase and decrease into lists. Afterward, I took the max for increases and the min for decreases.



```python
list_inc = []
list_inc_max = []
list_dec = []
list_dec_min = []
for i in range(mta_trend.shape[0]):
    j=1
    for j in range(mta_trend.shape[1]):
        if mta_trend[i,j]/mta_trend[i,j-1]-1 > 0:
            list_inc.append(mta_trend[i,j]/mta_trend[i,j-1]-1)
        else:
            list_dec.append(mta_trend[i,j]/mta_trend[i,j-1]-1)
    list_inc_max.append(max(list_inc))
    list_dec_min.append(min(list_dec))

         
for number in range(len(list_inc_max)):
    if list_inc_max[number] == max(list_inc_max):
        print("The number of ridership the steepest increase happened:", number+1)

for number in range(len(list_dec_min)):
    if list_dec_min[number] == min(list_dec_min):
        print("The number of ridership the steepest decrease happened:", number+1)
```

### Comment: Accordingly, we observe the steepest increase and the steepest decrease both occured for the ridership number 23
